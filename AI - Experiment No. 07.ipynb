## **Aim :	Implement Adaboost Ensemble Learning Algorithm for The Restaurant Waiting Problem.**

```python
class Perceptron : # With 2 inputs and 1 output
    def AdaBoost(self, examples, K):
        w = []        
        N = len(examples[1])
        y = examples[1]

        for i in range (0, N):            
            w.append(1/N)
        print("\nOriginal w = ", w)

        h = []

        for k in range (0,K) :
            print ("\n---------------------------------------------------\n")
            print("\nAs K = ", k+1)
            h.append([])
            h = self.L(examples,w)
            error = 0

            for j in range(0, N) :
                
                if( h[j]  != y[j] ):
                    error = error + w[j]
            print("\nError : ", error)

            for j in range(0, N) :
                if( h[j]  == y[j] ):                    
                    w[j] = w[j] * error / (1-error)
                    self.normalize(w)
            print("\nUpdated w = ", w)

        print ("\n---------------------------------------------------\n")
        print("\nResult using Final w : ")
        h = self.L(examples, w)
        print("\n")
        print(h)
         
        
    def normalize(self, w): 
            for t in range (0,len(w)) :                
                normalizer = 1 / float( sum(w) )

                w = [x * normalizer for x in w]

                
    def L(self, ex, w) : # Learning function
        hresult= []
        for i in range(0 , len(ex[1])):
            hresult.append(0)            
            hresult[i] = hresult[i] + ( w[i]*ex[0][i] ) 
        return hresult
```

```python
custn = int(input("Enter How Many Customers are in the Restaurant? [e.g. 5] : \t"))
queue = []
waittime = []
print("\nEnter Estimated Waiting Time in Minutes for Each Customer.")
for i in range(custn):
    queue.append(i + 1)
    print("\nCustomer No.", i + 1)
    time = int(input("Estimated Waiting Time: \t"))
    waittime.append(time)
```

```python
# the first list x-vector indicates a customer's number in the queue of the waiting list say 1,2...
# the second list y-vector indicates amount in minutes the customer will have to wait for the corresponding queue number

problem = [queue, waittime]

K = int(input("Enter Total Number of Decision Trees You want: \t"))

print ("\nK = ", K)
```

```python
p = Perceptron()

p.AdaBoost(problem, K)
```
