## **Aim :	Implement Naive Bayes Learning Algorithm for The Restaurant Waiting Problem.**

```python
class NaiveBayes:

    def __init__(self, f, r):
        self.features = f
        self.response = r

    def predict(self,custcase):
        anskeys = list(self.response.keys())
        ansvalues = dict.fromkeys(anskeys,0)
        print("\nCustomerCase : ", custcase)
        for key in anskeys :
            ansvalues[key] = self.response[key]
            
            for ikey, ival in custcase.items() :
                ansvalues[key] = ansvalues[key] * self.features[ikey][ival][key]

        print("\nPrediction Rates : ", ansvalues)

        #calculating MAP
        maxkey=""
        maxans=-1
        for ikey, ival in ansvalues.items():
            if ival > maxans :
                maxans= ival
                maxkey = ikey
        return maxkey
```

```python
#precalculated values from worksheet - "naive bayes classifier working"

response = {"Wait":0.4, "Leave":0.6}

features = {
    "Reservation":
                    {
                        "Yes" : {"Wait":0.5, "Leave":0.666667},
                         "No" : {"Wait":0.5, "Leave":0.333333}
                    } ,
    "Time>30":
                    {
                        "Yes" : {"Wait":0.25, "Leave":0.83333},
                         "No" : {"Wait":0.75, "Leave":0.16667}
                    }                  
}
```

```python
nb = NaiveBayes(features, response)

print("\nProbabilities :")
print("\nIF Reservation is : Booked and Taking Time<30 THEN : Wait.")
print("Here Probability : ", nb.features["Reservation"]["Yes"]["Wait"])
print("\nIF Reservation is : No and Taking Time>30 THEN : Leave.")
print("Here Probability : ", nb.features["Time>30"]["No"]["Leave"])
```

```python
print ("\n---------------------------------------------------\n")
resstatus = input("\nManager asks Customer, Have you reserved table?(Yes/No): ")
timestatus = input("\nNow Customer asks Manager, Will it take more than 30 mins?(Yes/No): ")
```

```python
custcase = {"Reservation":resstatus, "Time>30":timestatus}

print("\nManager predicts that Customer will : " , nb.predict(custcase) )
```
